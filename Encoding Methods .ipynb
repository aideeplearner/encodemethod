# Ordinal Encoding 

```python
#the ordinal encoding is more suitable for the non linear machine learning algorithms 
#in these we try to replace the categories with the arbitary numbers between 1 to k-1
#where k is the number of the columns in the data set
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal=OrdinalEncoder()
```

```python
x=ordinal.fit_transform(cat)
```

```python
final_cat=pd.DataFrame(x,columns=cat.columns)
```

```python
final_cat.head()
```
